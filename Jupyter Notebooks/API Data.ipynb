```python
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
df = pd.read_json("https://api.luftdaten.info/static/v2/data.dust.min.json")
df = pd.io.json.json_normalize(df.to_dict(orient='record')) \
               .set_index("id")
df
```

```python
val_df = (df["sensordatavalues"]
              .apply(pd.Series)
              .stack()
              .apply(pd.Series)
              .reset_index(level=1, drop=True)
              .rename(columns={"id": "value_id"}))
val_df
```

```python
meas_df = (pd.merge(left=val_df, right=df, left_index=True, right_index=True)
             .drop(columns=["sensordatavalues"]))
meas_df
```
